# 2. Data Cleaning and EDA

## 2.1 Dealing with null values

```python
# Calculate the number of null value and percentage for each features.
null_count  = pd.DataFrame(train_df.isna().sum(), columns = ['NaN'])
null_count['null_percentage (%)'] = (null_count['NaN']/ train_df.shape[0] * 100).round(2)

null_values = null_count[null_count['NaN'] >= 1]
null_values['null_percentage (%)'].sort_values(ascending=False)
```

#### Comments:
null values exist in the data. Drop 'Pool QC' , 'Misc Feature' , 'Alley', 'Fence', since the missing values are more than 80% of total dataset.

```python
# Drop 'Pool QC' , 'Misc Feature' , 'Alley', 'Fence'

train_df = train_df.drop(['Pool QC' , 'Misc Feature' , 'Alley', 'Fence'], axis=1)
test_df = test_df.drop(['Pool QC' , 'Misc Feature' , 'Alley', 'Fence'], axis=1)
```

### 2.1.1 Dealing with categorical null values

```python
null = train_df.columns[train_df.isnull().any()] # identify columns with null.

# extract categorical variables from train_df
cat_null_column = train_df[null].select_dtypes(include=['object']) 
cat_null_column.dtypes # verify
```

```python
# View the unique values in categorical dataframe with null values.

for a in cat_null_column:
    fig = plt.figure(figsize = (6,5))
    sns.countplot(x = a, data= cat_null_column, palette='Set3')
    
plt.show()
```

### 2.1.1.1 Examine null values for 'Fireplace Qu' and 'Garage'

Features with Null values consist of 'Fireplace Qu', 'Garage Type', 'GarageFinish', 'GarageQual', 'GarageCond',  Based on Data Dictionary, Nan means no fireplace/garage is built for the house. <br>
Use 'None' for the categorical variables. 

```python
# Fill Nan with 'None'
for col in ('Fireplace Qu', 'Garage Type', 'Garage Finish', 'Garage Qual', 'Garage Cond'):
    train_df[col] = train_df[col].fillna('None')
    test_df[col] = test_df[col].fillna('None')
```

```python
# Examine relationship between 'Garage Yr Blt' vs 'Year Built'

plt.figure(figsize=(8,8))
sns.scatterplot(data=train_df, x="Year Built", y="Garage Yr Blt")
plt.title('Year Built vs Garage Year Built', fontsize=20)
plt.xlabel('Year Built',fontsize=17)
plt.ylabel('Garage Year Built', fontsize=17)
plt.show()
```

```python
# Calculate the percentage of garage and house built on the same year
(train_df[train_df['Year Built'] == train_df['Garage Yr Blt']][['Year Built','Garage Yr Blt']].count()/train_df.shape[0] * 100).round(2)
```

#### Comments:
Majority of the Garage is built within the same year of the house built. Safe to use 'Year Built' to replace null values for 'Garage Yr Blt'.

```python
# fill missing value in 'Garage Yr Blt' = 'Year Built'
train_df['Garage Yr Blt'] = train_df['Garage Yr Blt'].fillna(train_df['Year Built'])
test_df['Garage Yr Blt'] = test_df['Garage Yr Blt'].fillna(test_df['Year Built'])
```

```python
# Verify missing value in 'Garage Area'
train_df[train_df['Garage Area'].isnull()][['Garage Type','Garage Cond', 'Garage Qual', 'Garage Finish', 'Garage Cars', 'Garage Area' ]]
```

Eventhough the Garage Type for the house index 1712 is Detached from home. However the other features has no relavent info about the garage. Assume there is no garage. Use 0 for 'Garage Cars' and 'Garage Area'.

```python
# fill missing value with 0
for col in ('Garage Cars', 'Garage Area'):
    train_df[col] = train_df[col].fillna(0)
    test_df[col] = test_df[col].fillna(0)
```

### 2.1.2 Dealing with numerical null values
Examine null value for 'LotFrontage' : Linear feet of street connected to property features

```python
# Use IterativeImputer to impute the missing data in Lot Frontage.

# import libraries
from sklearn.experimental import enable_iterative_imputer 
from sklearn.impute import IterativeImputer
from sklearn.linear_model import LinearRegression
```

```python
# Plot the distribution of 'Lot Frontage'

plt.figure(figsize=(6,6))
sns.displot(data= train_df, x="Lot Frontage")
plt.axvline(x=train_df['Lot Frontage'].mean(),
            color='red')
plt.title('Lot frontage distribution', fontsize=20)
plt.xlabel('Lot Frontage',fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.show()
```

```python
# extract numerical variables from train_df and test_df

numerical_train_df = train_df.select_dtypes(exclude=['object'])
numerical_train_df = numerical_train_df.drop(columns='SalePrice', axis = 1) #left out SalePrice so that IterativeImputer can transform test dataframe.
numerical_test_df = test_df.select_dtypes(exclude=['object'])
```

```python
# instantiate IterativeImputer
imp = IterativeImputer(estimator = LinearRegression())

# apply linear regression model imputation to missing value
it_num_train_df = imp.fit_transform(numerical_train_df)
it_num_train_df


it_num_test_df = imp.transform(numerical_test_df)
```

```python
# convert to table
numerical_train_df = pd.DataFrame(it_num_train_df, columns = numerical_train_df.columns)
numerical_test_df = pd.DataFrame(it_num_test_df, columns = numerical_test_df.columns)
```

```python
# replace train_df & test_df with imputed value.

train_df['Lot Frontage'] = numerical_train_df['Lot Frontage']
test_df['Lot Frontage'] = numerical_test_df['Lot Frontage']
```

```python
# Check the distribution after imputed missing value. 

plt.figure(figsize=(6,6))
sns.displot(data= numerical_train_df, x="Lot Frontage")
plt.axvline(x=numerical_train_df['Lot Frontage'].mean(),
            color='red')
plt.title('Lot frontage distribution', fontsize=20)
plt.xlabel('Lot Frontage',fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.show()
```

#### Comments:
The distribution seems identical after imputer. Imputer null values can be accepted.

### 2.3 Dealing with outliers

An outlier is observed for 'Garage Yr Blt' vs 'SalePrice. There is a data point where garage is built in about year 2200 which is not possible. Solution: Replace the outlier equal to 'Year Built'
