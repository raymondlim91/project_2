## 1. Import Train & Test dataset

```python
# Import Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_df = pd.read_csv("../datasets/train_cleaned.csv") # read train_cleaned csv file

print(train_df.shape)
train_df.head()
```

'MS SubClass', 'Overall Qual', 'Overall Cond' should be categorical variable.

### 4.1 Check for multicollinearity

```python
num_train_df.columns
```

Check multicollinearity for square feet (SF) of houses

```python
# Declare a function to create table for VIF based on features
def vif(features):
    vif = pd.DataFrame()
    vif['VIF'] = [variance_inflation_factor(features.values, i) for i in range(features.shape[1])]
    vif['featuers'] = features.columns
    return vif
```

```python
# import VIF libraries
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

### 6.1 OneHotEncoding for categorical variables.

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
concat_df = pd.concat([cat_train_df, cat_test_df], axis=0) #combine cat_train_df and cat_test_df
```

Found out there is some features in cat_test_df not found in cat_train_df. If onehotencode cat_train_df, the model will not be able to transform cat_test_df. <br>
There is a parameter in OneHotEncoder to set `handle_unknown = 'ignore'`, However it cannot be used together with `drop='first`. Not using `drop='first'` will produce extra 31 columns which creates unwanted noise to the model. <br>
To work around, combine cat_train_df and cat_test_df to do one hot encoding in one dataset.

```python
enc = OneHotEncoder(drop ='first') # create instance of one-hot-encoder 

enc.fit(concat_df) # fit 

encoded_train_df = pd.DataFrame(enc.transform(cat_train_df).toarray()) #transform
encoded_test_df = pd.DataFrame(enc.transform(cat_test_df).toarray()) #transform
```

```python
encoded_train_df.head()
```

```python
encoded_test_df.head()
```

### 7.1 Combine numerical and categorical X features

```python
X_train_processed = pd.merge(num_train_df_cleaned, encoded_train_df,left_index=True,right_index=True, how='inner')

X_train_processed.head()
```

```python
X_train_processed['SalePrice']
```

```python
X_test_processed = pd.merge(num_test_df_cleaned, encoded_test_df,left_index=True,right_index=True, how='inner')

X_test_processed.head()
```

```python
# export cleaned data to csv file
X_train_processed.to_csv(r'../datasets/train_processed.csv', index = False)
X_test_processed.to_csv(r'../datasets/test_processed.csv', index = False)
```
